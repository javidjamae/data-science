```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# General Exponential Function

### Overview

$ y = a^x $

### Plots

$y = a^x$ given $ \{ a > 1 \} $ 

```python
a = 2
x = np.linspace( -3, 10, 50, dtype=float )  
y = a ** x 

plt.plot( x, y )  
plt.show()
```

$y = a^x$ given $\{ a = 1 \}$

```python
a = 1
x = np.linspace( -3, 10, 50, dtype=float )  
y = a ** x 

plt.plot( x, y )  
plt.show()
```

$y = {a^x}$ given $\{ 0 < a < 1 \} $

```python
a = 0.5
x = np.linspace( -3, 10, 50, dtype=float )  
y = a ** x 

plt.plot( x, y )  
plt.show()
```

# Logarithmic Function

### Overview

The logarithmic function is:

$$y = {log_a}(x)$$

This function is the inverse of:

$$y = a^x$$

For example, if we have:

$$42^3 = 74088$$

Then we could calculate the exponent:

$$log_{42}(74088) = 3$$

### Calculating Using `math.log`

```python
import math
number = 42 ** 3
base = 42
exponent = math.log( number, base )

print( exponent )
```

### Calculating Using `numpy.log`

You can use the [logarithm change of base rule](https://en.wikipedia.org/wiki/Logarithm#Change_of_base) to convert the log calculation into a ratio of logs with a standard base.

$$\log_b x =  \frac{\log_k x}{\log_k b}$$

So, we can then use a ratio of base $e$ or base $10$ log to calculate:

$$\log_b x = \frac{\log_{10} x}{\log_{10} b} = \frac{\log_{e} x}{\log_{e} b}$$


```python
# https://stackoverflow.com/questions/25169297/numpy-logarithm-with-base-n
# https://en.wikipedia.org/wiki/Logarithm#Change_of_base
number = 42 ** 3
base = 42

exponent = np.log( number ) / np.log( base )
print( exponent )
```

### Plots

$y = {log_a}(x)$ given $\{ a > 1 \}$

```python
base = 2
x = np.linspace( .01, 10, 100, dtype=float )  
y = np.log( x ) / np.log( base )

plt.plot( x, y )  
plt.show()
```

$y = {log_a}(x)$ given $\{ a = 1 \}$

```python
base = 1
x = np.linspace( .01, 10, 100, dtype=float )  
y = np.log( x ) / np.log( base )
 
plt.plot( x, y )  
plt.show()
```

$y = {log_a}(x)$ given $\{ 0 < a < 1 \}$

```python
base = 0.5
x = np.linspace( .01, 10, 100, dtype=float )  
y = np.log( x ) / np.log( base )
 
plt.plot( x, y )  
plt.show()
```
