```python
import matplotlib.pyplot as plt
# Same as: 
# from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import warnings

warnings.filterwarnings( 'ignore' )
plt.rcParams[ 'figure.figsize' ] = ( 8, 6 )

stats = pd.read_csv( '../data/Birth-Internet-Income-Per-Country.csv' )
stats.columns = [ 'CountryName', 'CountryCode', 'BirthRate', 'InternetUsers', 'IncomeGroup' ]

movies = pd.read_csv('../data/Movie-Ratings.csv')
movies.columns = [ 'Film', 'Genre', 'CriticRating', 'AudienceRating', 'Budget', 'Year' ]
```

# boxplot()

```python
vis = sns.boxplot(data=stats, x="IncomeGroup", y="BirthRate")
```

# distplot()

```python
vis = sns.distplot(stats.InternetUsers)
```

**bins**

```python
vis = sns.distplot(stats.BirthRate, bins=50)
```

References:
- [Docs](https://seaborn.pydata.org/generated/seaborn.distplot.html)

# jointplot()

See the scatterplot and the individual distributions of a bivariate distribution.

```python
vis = sns.jointplot( data=movies, x='CriticRating', y='AudienceRating' )
```

**kind='hex'**

```python
vis = sns.jointplot( data=movies, x='CriticRating', y='AudienceRating', kind='hex' )
```

**kind='kde'**

```python
vis = sns.jointplot( data=movies, x='AudienceRating', y='CriticRating', kind='kde' )
```

# lmplot()

```python
vis = sns.lmplot( data=stats, x="InternetUsers", y="BirthRate" )
```

**aspect - define the aspect ratio for the plot**

```python
vis = sns.lmplot( data=stats, x="InternetUsers", y="BirthRate", fit_reg=False, aspect=3 )
```

**fit_reg - removes the regression line**

```python
vis = sns.lmplot( data=stats, x="InternetUsers", y="BirthRate", fit_reg=False )
```

**hue - define a separate variable to segment on for selecting colors**

```python
vis = sns.lmplot( data=stats, x="InternetUsers", y="BirthRate", fit_reg=False, hue="IncomeGroup" )
```

**color**

The lmplot uses a FacetGrid, so it doesn't pick up the size from `plt.rcParams[ 'figure.figsize' ]` like other plots. You have to specify a size parameter instead.

```python
vis = sns.lmplot( data=stats, x="InternetUsers", y="BirthRate", fit_reg=False, hue="IncomeGroup", size=8 )
```

**scatter_kws - pass through args to matplotlib**

```python
vis = sns.lmplot( data=stats, x="InternetUsers", y="BirthRate", fit_reg=False, hue="IncomeGroup", scatter_kws={ 's': 150 } )
```
