# Overview

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

array = np.arange( 23, 50 )
```

```python
print( array )
```

### Basic Plot

```python
plt.plot( array )
plt.show()
```

# Plot Style

### Plot Title

```python
plt.plot( array )
plt.title( 'Title' )
plt.show()
```

### Axis Labels

```python
plt.plot( array )
plt.xlabel( 'X-Label' )
plt.ylabel( 'Y-Label' )
plt.show()
```

### Plot Size

```python
plt.rcParams['figure.figsize'] = 15, 8
plt.plot( array, c="Black" )
plt.show()

plt.rcParams['figure.figsize'] = 6, 4 # Reset for future plots in this notebook
```

### X-axis Labels

**Custom range**

```python
plt.plot( np.arange( 0, 5 ) )
plt.xticks( list( range( 0, 10, 3 ) ) ) # Label with 0 - 9, by 3's
plt.show()
```

**Lookup range from another array**

```python
labels = [ 'one', 'two', 'three', 'four', 'five' ]
plt.plot( np.arange( 0, 5 ) )
plt.xticks( list( range( 0, 5 ) ), labels )
plt.show()
```

**Vertical Labels**

```python
labels = [ 'one', 'two', 'three', 'four', 'five' ]
plt.plot( np.arange( 0, 5 ) )
plt.xticks( list( range( 0, 5 ) ), labels, rotation='vertical' )
plt.show()
```

# Legend

```python
plt.plot( array, label="first" )
plt.legend()
plt.show()
```

# Line Style 

### Line Color

```python
plt.plot( array, c="Black" )
plt.show()
```

### Line Style

```python
plt.plot( array, ls="--" )
plt.show()
```

### Marker

You can use `marker` to define the marker style and `ms` to define the marker size.

```python
plt.plot( array, marker="^", ms=12 ) # triangle
# plt.plot( array, marker="d" ) # diamond
# plt.plot( array, marker="s" ) # square
plt.show()
```
