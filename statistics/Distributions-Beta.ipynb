### Haldane's Prior - Beta( 0, 0 )

Another *uninformative prior* is the Beta(0, 0) prior (the [Haldane's prior probability](https://en.wikipedia.org/wiki/Beta_distribution#Haldane.27s_prior_probability_.28Beta.280.2C0.29.29)) where both $\alpha=0$ and $\beta=0$. This doesn't produce a "proper" prior distribution, per se, as it has infinite mass at 0 and 1, because the $0$'s will cause $1/0$ terms in the Beta function. The Beta(0, 0) can't be graphed easily (as shown below). 

But, if you want an uninformative prior, since you just need to do some basic arithmetic to calculate the updated $\alpha$ and $\beta$ terms for your posterior distribution as soon as you collect some data for each variation, it doesn't really matter much what the initial prior distribution looks like.

```python
a = 0
b = 0
x = np.arange( 0, 1, 0.01 )
y = beta.pdf( x, a, b )
plt.plot( x, y )
plt.show()
```
